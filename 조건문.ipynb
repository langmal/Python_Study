## 조건문

```python
A, B = input().split()

A = int(A)
B = int(B)

if A > B :
  print('>')
elif A < B :
  print('<')
else :
  print('==')
```

```python
score = int(input())

if 90<=score<=100 :
  print('A')
elif 80<=score<=89 :
  print('B')
elif 70<=score<=79 :
  print('C')
elif 60<=score<=69 :
  print('D')
else :
  print('F')
```

```python
year = int(input())

if year % 4 == 0 and (year % 100 != 0 or year % 400 == 0):
  print("1")
else :
  print("0")
```

```python
x = int(input())
y = int(input())

if x>0 and y>0 :
  print('1')
elif x<0 and y>0 :
  print('2')
elif x<0 and y<0 :
  print('3')
else :
  print('4')
```

```python
H, M = input().split()
H = int(H)
M = int(M)

if M < 45 :
  if H >= 1 :
    print(H-1, 60+(M-45))
  else :
    print(24+(H-1), 60+(M-45))
else :
  print(H, M-45)
```

```python
A, B = input().split()
Timer = int(input())
A = int(A)
B = int(B)

H = A + Timer//60 
M = B + Timer%60

if M >= 60 :
  H += 1
  M -= 60
if H >= 24 :
  H -=24

print(H, M)
```

```python
A, B, C = input().split()
A = int(A)
B = int(B)
C = int(C)
prize = 0

Max = (max(A, B, C))

if A == B == C :
  prize = 10000 + A*1000
elif A == B or A ==C : 
  prize = 1000 + A*100
elif B == C :
  prize = 1000 + B*100
else :
  prize = Max*100

print(prize)
```
